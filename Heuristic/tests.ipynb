```python
from copy import deepcopy
import random

def shake(schedule, Max_d):
    temp_schedule = deepcopy(schedule)
    
    # Wähle eine zufällige Nachbarschaftsoperation: N1, N2 oder N3
    operation = random.choice(['N1', 'N2', 'N3'])
    keys = list(schedule.keys())
    if operation == 'N3' and len(keys)>1:
        # Randomly select two doctors and one patient from each for N3
        key1, key2 = random.sample(keys, 2)
        index1 = random.randint(0, temp_schedule[key1].shape[0] - 1)
        index2 = random.randint(0, temp_schedule[key2].shape[0] - 1)
        temp_schedule = nh.swap_patients_between_docs_N3(temp_schedule, [key1, index1], [key2, index2],  Max_d=Max_d)

    elif operation == 'N1':
        # Randomly select a doctor and two patient indices for N1
        key = random.choice(keys)
        if temp_schedule[key].shape[0] > 1:
            i, j = random.sample(range(temp_schedule[key].shape[0]), 2)
            temp_schedule = nh.swap_patients_by_doc_N1(temp_schedule, key, i, j, Max_d=Max_d)
    
    elif operation == 'N2':
        # Randomly select a doctor and two patient indices for N2
        key = random.choice(keys)
        if temp_schedule[key].shape[0] > 1:
            i, j = random.sample(range(temp_schedule[key].shape[0]), 2)
            patient_id_to_move = temp_schedule[key][i, 0]
            patient_id_target = temp_schedule[key][j, 0]
            temp_schedule = nh.insert_patient_by_doc_N2(temp_schedule, key, patient_id_to_move, patient_id_target,  Max_d=Max_d)
    


    overall_tardiness = 0
    for key in temp_schedule:
        overall_tardiness += nh.weighted_tardiness_per_doc(temp_schedule, key)
    
    return temp_schedule, overall_tardiness

def VND(schedule, overall_tardiness, neighborhoods, Max_d=120):
    current_schedule = deepcopy(schedule)
    current_tardiness = overall_tardiness
    improvement = True

    while improvement:
        improvement = False
        for neighborhood in neighborhoods:
            # Führe die lokale Suche für die aktuelle Nachbarschaft aus
            if neighborhood == "N1":
                new_schedule, _ = nh.local_search(schedule, neighborhood="N1", Max_d=Max_d)
            elif neighborhood == "N2":
                new_schedule, _ = nh.local_search(schedule, neighborhood="N2", Max_d=Max_d)
            elif neighborhood == "N3":
                new_schedule, _ = nh.local_search(schedule, neighborhood="N3", Max_d=Max_d)
            else:
                continue

            # Vergleiche die neue und aktuelle Lösung
            if _ < current_tardiness:
                current_schedule = new_schedule
                current_tardiness = _
                improvement = True
                break  # Beginne die Suche erneut bei der ersten Nachbarschaft

    return current_schedule, current_tardiness

import time

def general_vns(initial_schedule, initial_tardiness ,neighborhoods, Max_d=120, no_improvement_limit=10, time_limit=90):
    current_schedule = deepcopy(initial_schedule)
    best_schedule = current_schedule
    best_schedule_tardiness = initial_tardiness
    no_improvement_count = 0

    start_time = time.time()


    while time.time() - start_time <= time_limit:
        # Shake: Erzeugen einer neuen Lösung durch eine zufällige Veränderung
        shaken_schedule, shaken_tardiness = shake(current_schedule, Max_d=Max_d)

        # VND: Anwenden der lokalen Optimierung
        vnd_schedule, vnd_tardiness = VND(shaken_schedule, shaken_tardiness ,neighborhoods, Max_d)

        # Bewertung der neuen Lösung
        if vnd_tardiness < best_schedule_tardiness:
            best_schedule = vnd_schedule
            best_schedule_tardiness = vnd_tardiness
            current_schedule = vnd_schedule
            no_improvement_count = 0  # Reset, da eine Verbesserung gefunden wurde
        else:
            no_improvement_count += 1

    return best_schedule, best_schedule_tardiness
```

```python
from copy import deepcopy
import Neighbourhoods as nh

def best_swap_N1(schedule, key):
    # Anzahl der Patienten bei diesem Doktor
    num_patients = schedule[key].shape[0]
    
    # Initialisiere die Variablen für die beste Lösung
    best_schedule = deepcopy(schedule)
    lowest_tardiness = nh.weighted_tardiness_per_doc(schedule, key)
    
    # Durchlaufe alle möglichen Paare von Patienten, vermeide doppelte Tauschoperationen
    for i in range(num_patients):
        for j in range(i + 1, num_patients):
            # Führe den Tausch durch
            temp_schedule = deepcopy(schedule)
            temp_schedule = nh.swap_patients_by_doc_N1(temp_schedule, key, i, j)
            
            # Berechne den weighted tardiness für den getauschten Schedule
            tardiness = nh.weighted_tardiness_per_doc(temp_schedule, key)
            
            # Aktualisiere die beste Lösung, wenn ein niedrigerer Wert gefunden wird
            if tardiness < lowest_tardiness:
                print(f"Bessere Lösung gefunden: Tausch von {i} und {j}, von {lowest_tardiness} auf {tardiness}")
                lowest_tardiness = tardiness
                best_schedule = temp_schedule
    
    return best_schedule, lowest_tardiness

def best_insert_N2(schedule, key):
    # Anzahl der Patienten bei diesem Doktor
    num_patients = schedule[key].shape[0]
    
    # Initialisiere die Variablen für die beste Lösung
    best_schedule = deepcopy(schedule)
    lowest_tardiness = nh.weighted_tardiness_per_doc(schedule, key)
    
    # Iteriere über alle möglichen Paare von Patienten-IDs
    for i in range(num_patients):
        for j in range(num_patients):
            if i != j:
                # Bestimme die Patienten-IDs
                patient_id_to_move = schedule[key][i, 0]
                patient_id_target = schedule[key][j, 0]
                
                # Führe die Einfügeoperation durch
                temp_schedule = nh.insert_patient_by_doc_N2(schedule, key, patient_id_to_move, patient_id_target)
                
                # Berechne den weighted tardiness für den veränderten Schedule
                tardiness = nh.weighted_tardiness_per_doc(temp_schedule, key)
                
                # Aktualisiere die beste Lösung, wenn ein niedrigerer Wert gefunden wird
                if tardiness < lowest_tardiness:
                    print(f"Bessere Lösung gefunden: Insert von {i} vor {j}, von {lowest_tardiness} auf {tardiness}")
                    lowest_tardiness = tardiness
                    best_schedule = temp_schedule
    
    return best_schedule, lowest_tardiness


def best_swap_N3(schedule, key1, key2):
    # Anzahl der Patienten für die beiden Doktoren
    num_patients_key1 = schedule[key1].shape[0]
    num_patients_key2 = schedule[key2].shape[0]
    
    # Initialisiere die Variablen für die beste Lösung
    best_schedule = deepcopy(schedule)
    lowest_combined_tardiness = (
        nh.weighted_tardiness_per_doc(schedule, key1) +
        nh.weighted_tardiness_per_doc(schedule, key2)
    )
    
    # Durchlaufe alle möglichen Paare von Patienten zwischen den beiden Ärzten
    for i in range(num_patients_key1):
        for j in range(num_patients_key2):
            # Führe den Tausch durch
            swapped_schedule = nh.swap_patients_between_docs_N3(schedule, [key1, i], [key2, j])
            
            # Berechne die kombinierten weighted tardiness für beide Doktoren
            combined_tardiness = (
                nh.weighted_tardiness_per_doc(swapped_schedule, key1) +
                nh.weighted_tardiness_per_doc(swapped_schedule, key2)
            )
            
            # Aktualisiere die beste Lösung, wenn ein niedrigerer Wert gefunden wird
            if combined_tardiness < lowest_combined_tardiness:
                print(f"Bessere Lösung gefunden: Insert von {i} vor {j}, von {lowest_combined_tardiness} auf {combined_tardiness}")
                lowest_combined_tardiness = combined_tardiness
                best_schedule = swapped_schedule
    
    return best_schedule, lowest_combined_tardiness

def iterative_best_swap_N1(schedule, key):
    improved = True
    current_schedule = deepcopy(schedule)
    best_tardiness = nh.weighted_tardiness_per_doc(current_schedule, key)
    
    while improved:
        improved = False
        best_schedule, lowest_tardiness = best_swap_N1(current_schedule, key)
        
        if lowest_tardiness < best_tardiness:
            current_schedule = best_schedule
            best_tardiness = lowest_tardiness
            improved = True
    
    return current_schedule, best_tardiness

def iterative_best_insert_N2(schedule, key):
    improved = True
    current_schedule = deepcopy(schedule)
    best_tardiness = nh.weighted_tardiness_per_doc(current_schedule, key)
    
    while improved:
        improved = False
        best_schedule, lowest_tardiness = best_insert_N2(current_schedule, key)
        
        if lowest_tardiness < best_tardiness:
            current_schedule = best_schedule
            best_tardiness = lowest_tardiness
            improved = True
    
    return current_schedule, best_tardiness

def iterative_best_swap_N3(schedule, key1, key2):
    improved = True
    current_schedule = deepcopy(schedule)
    best_tardiness = (
        nh.weighted_tardiness_per_doc(current_schedule, key1) +
        nh.weighted_tardiness_per_doc(current_schedule, key2)
    )
    
    while improved:
        improved = False
        best_schedule, lowest_tardiness = best_swap_N3(current_schedule, key1, key2)
        
        if lowest_tardiness < best_tardiness:
            current_schedule = best_schedule
            best_tardiness = lowest_tardiness
            improved = True
    
    return current_schedule, best_tardiness
```

```python
import numpy as np

def feasibility_check(schedule, key, Max_d=120):
    # Extrahieren Sie die relevanten Spalten für den angegebenen Schlüssel
    fourth_column = schedule[key][:, 3]
    sixth_column = schedule[key][:, 5]

    # Berechnen Sie die Bedingung: Spalte 4 < Spalte 6 + Max_d
    condition = fourth_column < (sixth_column + Max_d)

    # Prüfen Sie, ob die Bedingung für mindestens eine Zeile erfüllt ist
    if np.any(condition):
        return False
    else:
        return True
```

```python
import random

def shake(schedule, keys):
    temp_schedule = deepcopy(schedule)
    
    # Wähle eine zufällige Nachbarschaftsoperation: N1, N2 oder N3
    operation = random.choice(['N1', 'N2', 'N3'])
    
    if operation == 'N1':
        # Randomly select a doctor and two patient indices for N1
        key = random.choice(keys)
        if temp_schedule[key].shape[0] > 1:
            i, j = random.sample(range(temp_schedule[key].shape[0]), 2)
            temp_schedule[key][[i, j]] = temp_schedule[key][[j, i]]
            nh.calculate_start_and_endtimes_per_doc(schedule=temp_schedule, key=key)
    
    elif operation == 'N2':
        # Randomly select a doctor and two patient indices for N2
        key = random.choice(keys)
        if temp_schedule[key].shape[0] > 1:
            i, j = random.sample(range(temp_schedule[key].shape[0]), 2)
            patient_id_to_move = temp_schedule[key][i, 0]
            patient_id_target = temp_schedule[key][j, 0]
            temp_schedule = nh.insert_patient_by_doc_N2(temp_schedule, key, patient_id_to_move, patient_id_target)
    
    elif operation == 'N3':
        # Randomly select two doctors and one patient from each for N3
        if len(keys) > 1:
            key1, key2 = random.sample(keys, 2)
            index1 = random.randint(0, temp_schedule[key1].shape[0] - 1)
            index2 = random.randint(0, temp_schedule[key2].shape[0] - 1)
            temp_schedule = nh.swap_patients_between_docs_N3(temp_schedule, [key1, index1], [key2, index2])
    
    return temp_schedule

# def VNS_with_random_shake(schedule, max_iterations=100):
#     current_schedule = deepcopy(schedule)
#     keys = list(current_schedule.keys())
    
#     current_tardiness = sum(
#         nh.weighted_tardiness_per_doc(current_schedule, key) for key in keys
#     )
#     iteration = 0

#     while iteration < max_iterations:
#         iteration += 1
#         print(f"Iteration: {iteration}")
        
#         # Shake: Wähle zufällig eine Operation aus N1, N2, oder N3
#         shaken_schedule = shake(current_schedule, keys)
        
#         # Lokale Suche in der geschüttelten Lösung
#         improved = False
#         for key in keys:
#             new_schedule, new_tardiness = iterative_best_swap_N1(shaken_schedule, key)
#             if new_tardiness < current_tardiness:
#                 shaken_schedule = new_schedule
#                 current_tardiness = new_tardiness
#                 improved = True
                
#             new_schedule, new_tardiness = iterative_best_insert_N2(shaken_schedule, key)
#             if new_tardiness < current_tardiness:
#                 shaken_schedule = new_schedule
#                 current_tardiness = new_tardiness
#                 improved = True
        
#         for key1, key2 in combinations(keys, 2):
#             new_schedule, new_tardiness = iterative_best_swap_N3(shaken_schedule, key1, key2)
#             if new_tardiness < current_tardiness:
#                 shaken_schedule = new_schedule
#                 current_tardiness = new_tardiness
#                 improved = True
        
#         # Aktualisiere die aktuelle Lösung, wenn eine bessere gefunden wurde
#         if improved:
#             current_schedule = shaken_schedule

#     return current_schedule, current_tardiness

# # Beispiel der Ausführung
# temp_schedule = deepcopy(result[0])
# final_schedule, final_tardiness = VNS_with_random_shake(temp_schedule, max_iterations=10)
# print("Finale Tardiness:", final_tardiness)

def VNS_with_random_shake(schedule, max_iterations=100, max_no_improve_iterations=10):
    current_schedule = deepcopy(schedule)
    keys = list(current_schedule.keys())
    
    current_tardiness = sum(
        nh.weighted_tardiness_per_doc(current_schedule, key) for key in keys
    )
    iteration = 0
    no_improve_iterations = 0

    while iteration < max_iterations and no_improve_iterations < max_no_improve_iterations:
        iteration += 1
        
        # Shake: Wähle zufällig eine Operation aus N1, N2, oder N3
        shaken_schedule = shake(current_schedule, keys)
        
        # Lokale Suche in der geschüttelten Lösung
        improved = False
        for key in keys:
            new_schedule, new_tardiness = iterative_best_swap_N1(shaken_schedule, key)
            if new_tardiness < current_tardiness:
                shaken_schedule = new_schedule
                current_tardiness = new_tardiness
                improved = True
                
            new_schedule, new_tardiness = iterative_best_insert_N2(shaken_schedule, key)
            if new_tardiness < current_tardiness:
                shaken_schedule = new_schedule
                current_tardiness = new_tardiness
                improved = True
        
        for key1, key2 in combinations(keys, 2):
            new_schedule, new_tardiness = iterative_best_swap_N3(shaken_schedule, key1, key2)
            if new_tardiness < current_tardiness:
                shaken_schedule = new_schedule
                current_tardiness = new_tardiness
                improved = True

        # Überprüfe, ob eine Verbesserung stattgefunden hat
        if improved:
            current_schedule = shaken_schedule
            no_improve_iterations = 0  # Setze den Zähler für keine Verbesserung zurück
        else:
            no_improve_iterations += 1  # Erhöhe den Zähler für keine Verbesserung

    return current_schedule, current_tardiness

# Beispiel der Ausführung
temp_schedule = deepcopy(result[0])
final_schedule, final_tardiness = VNS_with_random_shake(temp_schedule, max_iterations=10, max_no_improve_iterations=10)
print("Finale Tardiness:", final_tardiness)
```
